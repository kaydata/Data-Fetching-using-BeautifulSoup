## Getting Data using BeautifulSoup

```python
# Fetch the HTML content
url = 'https://www.epra.go.ke/services/petroleum/petroleum-prices/'
response = requests.get(url)
response.raise_for_status()  # Check for a valid response

# Parse the HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Locate the file links
for link in soup.find_all('a', href=True):
    href = link['href']
    if href.endswith('.csv') or href.endswith('.xlsx'):
        file_url = urljoin(url, href)
        print(f"Fetching file from {file_url}")
        
        # Download the file
        file_response = requests.get(file_url)
        file_response.raise_for_status()
        
        # Determine the file name from the URL
        file_name = file_url.split('/')[-1]
        
        # Save the file to disk
        with open(file_name, 'wb') as file:
            file.write(file_response.content)
        
        # Read the file with pandas and print the first 5 rows
        if file_name.endswith('.csv'):
            df = pd.read_csv(file_name)
        else:
            df = pd.read_excel(file_name)
df.head()
```
